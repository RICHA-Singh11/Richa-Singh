# <span style='color:red'> DETECTION AND REMOVAL OF OUTLIERS: </span> 

```python
# Position of the Outlier
print(np.where(data['Sales']>15000))
```

```python
# Position of the Outlier
print(np.where(data['Discount'].value_counts()>0.70))
```

```python
# Position of the Outlier
print(np.where(data['Quantity']>10))
```

```python
# IQR
Q1 = np.percentile(data['Sales'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(data['Sales'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1
```

```python
print(IQR)
```

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1 - 1.5 * IQR
upper_bound=Q3 + 1.5 * IQR
print ("lower bound: ")
print(lower_bound)
print ("upper bound: ")
print (upper_bound)
```

```python
IQR = data[~((data < lower_bound) |(data > upper_bound)).any(axis=1)]
IQR.shape
```

```python
# Before removing outliers

sns.boxplot(data['Sales'])
```
